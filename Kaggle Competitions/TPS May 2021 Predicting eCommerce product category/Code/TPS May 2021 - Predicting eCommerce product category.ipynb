The following figure shows the distribution of each feature for both the train and test sets. All of the features are of integer type and we do not have the name of the feature so we do not know if they are categorical or numerical in nature.

```python
fig, axs = plt.subplots(nrows=17, ncols=3, figsize=(12, 50))
for feature, ax in zip(features, axs.ravel()):
    feature_prop_train = train.groupby(feature)[feature].count()/len(train)
    feature_prop_test = test.groupby(feature)[feature].count()/len(test)
    ax.bar(feature_prop_train.index, feature_prop_train.values, color='blue', alpha=0.3, label='Train')
    ax.bar(feature_prop_test.index, feature_prop_test.values, color='green', alpha=0.3, label='Test')
    ax.legend(loc='upper right')
    ax.set(xlabel=feature, ylabel='proportion')
fig.suptitle('Distribution of features', y=1, fontsize=20)
fig.tight_layout()
```

##### Takeaways:
1. The feature distributions are identical between the train and test sets.
2. In all of the features, the majority of the values are 0.
3. The distribution of all features is right-skewed.

## 1.3 Relationship of features with the target <a class='anchor' id='13'></a>
The following figure shows the proportion of the target classes for the values of the features.

```python
fig, axs = plt.subplots(nrows=17, ncols=3, figsize=(12, 70))
for i, (feature, ax) in enumerate(zip(features, axs.ravel())):
    df_plot = train.groupby([feature, target]).size().reset_index().rename(columns={0: 'count'})
    df_plot['sum'] = df_plot.groupby(feature)['count'].transform('sum')
    df_plot['prop'] = df_plot['count']/df_plot['sum']
    df_plot.drop(['count', 'sum'], axis=1, inplace=True)
    df_plot = df_plot.pivot(columns=target, index=feature, values='prop')
    
    df_plot.plot(kind='bar', stacked=True, ax=ax)
    ax.tick_params(axis='x', labelrotation=0)
    ax.set(ylabel='proportion')
    if i>=1:
        ax.legend('', frameon=False)
fig.suptitle('Proportion of target classes by feature values', y=1, fontsize=20)
fig.tight_layout()
```

##### Takeaways:
1. In the majority of the features, the proportion of the target classes changes when the feature values are high.

# 3 MODELING <a class='anchor' id='3'></a>

##### Function to train model and generate out-of-fold validation and test predictions
This function trains the model and generates out-of-fold validation and test predictions from cross-validation. Because the target is imbalanced, we use StratifiedKFold (with 5 splits and random shuffle).

```python
def train_and_predict(model, train, test, features, target, n_splits, fit_params={}, evaluation=True):

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=100)

    folds_score = []
    preds_test_list = []
    preds_test = np.zeros((len(test), 4))
    preds_val_list = []
    preds_val = np.zeros((len(train), 4))

    for fold, (train_idx, val_idx) in enumerate(skf.split(train[features], train[target])):
        X_train, y_train = train[features].iloc[train_idx], train[target].iloc[train_idx]
        X_val, y_val = train[features].iloc[val_idx], train[target].iloc[val_idx]

        model = model
        if evaluation==True:
            model.fit(X_train, y_train, eval_set=[(X_val, y_val)], **fit_params)
        else:
            model.fit(X_train, y_train, **fit_params)

        pred_val = model.predict_proba(X_val)
        preds_val_list.append(pred_val)
        preds_val[val_idx] = np.mean(preds_val_list, axis=0)

        preds_test_fold = model.predict_proba(test[features])
        preds_test_list.append(preds_test_fold)

        fold_score = log_loss(y_val, pred_val)
        folds_score.append(fold_score)

        print(f'FOLD {fold} SCORE: {fold_score}')
    print(f'MEAN CV SCORE: {np.mean(folds_score)}')
    print(f'OOF SCORE: {log_loss(train[target], preds_val)}')

    preds_test = np.mean(preds_test_list, axis=0)

    return preds_val, preds_test
```

##### Dictionaries to save validation and test predictions

```python
preds_val_dict = {}
preds_test_dict = {}
```

## 3.1 Logistic Regression <a class='anchor' id='31'></a>

### 3.1.1 Training and predicting <a class='anchor' id='311'></a>

```python
params_lr = {
          'random_state': 100,
          'penalty':'l2',
          'multi_class':'ovr',
          'solver':'lbfgs',  
          'C':0.01,
          'max_iter':10000 ,
          'class_weight':None        
          }

base_model = LogisticRegression(**params_lr)
model = CalibratedClassifierCV(base_model, method='sigmoid', cv=5)
preds_val_dict['lr'], preds_test_dict['lr'] = train_and_predict(model, train, test, features, target, n_splits=5, fit_params={}, evaluation=False)
```

## 3.2 Random Forest <a class='anchor' id='32'></a>

### 3.2.1 Training and predicting <a class='anchor' id='321'></a>

```python
base_model = RandomForestClassifier(random_state=100)
model = CalibratedClassifierCV(base_model, method='sigmoid', cv=5)

preds_val_dict['rf'], preds_test_dict['rf'] = train_and_predict(model, train, test, features, target, n_splits=5, fit_params={}, evaluation=False)
```

## 3.3 LightGBM <a class='anchor' id='33'></a>

### 3.3.1 Training and predicting <a class='anchor' id='331'></a>

```python
params_lgbm = {'random_state': 100,
                  'n_estimators': 50000,
                  'cat_l2': 1,
                  'cat_smooth': 41,
                  'colsample_bytree': 0.10152839180930642,
                  'learning_rate': 0.028561805912614473,
                  'max_depth': 77,
                  'min_child_samples': 256,
                  'min_data_per_group': 293,
                  'num_leaves': 69,
                  'reg_alpha': 41.076503175793036,
                  'reg_lambda': 0.0011579102445110068,
                  'subsample': 0.3473382674747556}

model = LGBMClassifier(**params_lgbm)

fit_params_lgbm = {'verbose': 200,
                   'early_stopping_rounds': 100}

preds_val_dict['lgbm'], preds_test_dict['lgbm'] = train_and_predict(model, train, test, features, target, n_splits=5, fit_params=fit_params_lgbm, evaluation=True)
```

## 3.4 CatBoost <a class='anchor' id='34'></a>

### 3.4.1 Training and predicting <a class='anchor' id='341'></a>

```python
params_cb = {'random_state': 100,
             'learning_rate': 0.03470328317940195, 
           'depth': 2, 
           'l2_leaf_reg': 820.7804346737378, 
           'random_strength': 0.336019499813798, 
           'border_count': 128,
           'grow_policy': 'Lossguide',
           'min_data_in_leaf': 267}

model = CatBoostClassifier(**params_cb)

fit_params_cb = {'verbose_eval': 50}

preds_val_dict['cb'], preds_test_dict['cb'] = train_and_predict(model, train, test, features, target, n_splits=5, fit_params=fit_params_cb, evaluation=True)
```
