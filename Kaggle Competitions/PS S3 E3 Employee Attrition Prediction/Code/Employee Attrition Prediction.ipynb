### Hypetesis tests of the difference in feature means between attrition and non-attrition.
The following figure shows the percentage difference betwwen the mean of each feature between churn and not churn, as well as an indication of wether the dirrence in means is statistically signifficant based on an hypotesis test of difference in means (t-test).

```python
ttests_df = pd.DataFrame()
for col in features_continuous + features_ordinal:
    ttest_var = ttest(train[train[target] == 1][col], train[train[target] == 0][col], correction = False)
    ttest_var['Mean Not Attrition'] = np.mean(train[train[target] == 0][col])
    ttest_var['Mean Attrition'] = np.mean(train[train[target] == 1][col])
    ttest_var['% Difference'] = (ttest_var['Mean Attrition'] - ttest_var['Mean Not Attrition']) / ttest_var['Mean Not Attrition'] * 100
    ttest_var['Feature'] = col
    ttests_df = pd.concat([ttests_df, ttest_var])
ttests_df.set_index('Feature', drop=True, inplace=True)
ttests_df = ttests_df[['Mean Not Attrition', 'Mean Attrition', '% Difference', 'p-val']]
ttests_df['p-val'] = ttests_df['p-val'].round(5)
ttests_df['pval_str'] = ttests_df['p-val'].astype(object).apply(lambda x: "{:.5f}".format(x))

conditions = [ttests_df['p-val'] <= 0.001,
             (ttests_df['p-val'] > 0.001) & (ttests_df['p-val'] <= 0.05),
             (ttests_df['p-val'] > 0.05) & (ttests_df['p-val'] <= 0.1)]
choices = [ttests_df['pval_str']+'***', ttests_df['pval_str']+'**', ttests_df['pval_str']+'*']
default = ttests_df['pval_str']


ttests_df['significant'] = np.where(ttests_df['p-val'] <= 0.1, 1, 0)
ttests_df['p-val'] = np.select(conditions, choices, default)
ttests_df.drop('pval_str', axis=1, inplace=True)
ttests_df.sort_values(by='% Difference', ascending=False, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.bar(ttests_df.index.values, ttests_df['% Difference'],
      color = np.where(ttests_df['significant'] == 1, 'black', 'gray'))
ax.set_xticklabels(ttests_df.index.values, rotation=90)
ax.set(xlabel='Feature', ylabel='% Difference', title='% Difference in means between attrition and not attrition')
legend_text = {'Significant at 10%': 'black', 'Not significant at 10%': 'gray'}         
labels = list(legend_text.keys())
handles = [plt.Rectangle((0,0),1,1, color=legend_text[label]) for label in labels]
ax.legend(handles, labels)
ax.grid(axis='x')
```

# 3 MODELING <a class='anchor' id='3'></a>

##### Function to train model and generate out-of-fold validation and test predictions
This function trains the model and generates out-of-fold validation and test predictions from cross-validation. Because the target is imbalanced, we use StratifiedKFold (with 5 splits and random shuffle).

```python
def train_and_predict(model, train, test, features, target, n_splits, fit_params={}, importances=False, evaluation=True):

    preds_val = np.zeros(train.shape[0])
    preds_test = np.zeros(test.shape[0])
    folds_score = []
    importances_list = []

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=100)
    for fold, (train_idx, val_idx) in enumerate(skf.split(train[features], train[target])):
        X_train, y_train = train[features].iloc[train_idx], train[target].iloc[train_idx]
        X_val, y_val = train[features].iloc[val_idx], train[target].iloc[val_idx]

        model = model
        if evaluation==True:
            model.fit(X_train, y_train, eval_set=[(X_val, y_val)], **fit_params)
        else:
            model.fit(X_train, y_train, **fit_params)

        pred_val = model.predict_proba(X_val)[:,1]
        preds_val[val_idx] = pred_val

        pred_test = model.predict_proba(test[features])[:,1]
        preds_test += pred_test / n_splits

        fold_score = roc_auc_score(y_val, pred_val)
        folds_score.append(fold_score)
        
        if importances==True:
            importances_list.append(model.coef_.ravel())

        print(f'FOLD {fold} AUC SCORE: {fold_score}')
    print(f'MEAN CV AUC SCORE: {np.mean(folds_score)}')
    print(f'OOF AUC SCORE: {roc_auc_score(train[target], preds_val)}')
    
    if importances==True:
        return preds_val, preds_test, importances_list
    else:
        return preds_val, preds_test
```

## 3.4 Stacking  <a class='anchor' id='34'></a>

```python
print(preds_val_df.head())
print(preds_test_df.head())
```

```python
## Preparing level 1 train and test sets for stacking
X_train_l1 = preds_val_df
y_train_l1 = train[target]

X_test_l1 = preds_test_df

## Training and predicting with level 2 metamodel
preds_val = np.zeros(X_train_l1.shape[0])
preds_test = np.zeros(X_test_l1.shape[0])
folds_score = []

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_l1, y_train_l1)):
    X_train, y_train = X_train_l1.iloc[train_idx], y_train_l1.iloc[train_idx]
    X_val, y_val = X_train_l1.iloc[val_idx], y_train_l1.iloc[val_idx]
    
    model = LogisticRegression(random_state=100)
    model.fit(X_train, y_train)
    
    pred_val = model.predict_proba(X_val)[:,1]
    preds_val[val_idx] = pred_val
    
    pred_test = model.predict_proba(X_test_l1)[:,1]
    preds_test += pred_test / 5
    
    fold_score = roc_auc_score(y_val, pred_val)
    folds_score.append(fold_score)
    
    print(f'FOLD {fold} SCORE: {fold_score}')
print(f'MEAN CV SCORE: {np.mean(folds_score)}')
print(f'OOF SCORE: {roc_auc_score(y_train_l1, preds_val)}')

## Puting stacking predictions in the dataframes of predictions
preds_val_df['stack'] = preds_val
preds_test_df['stack'] = preds_test
```
